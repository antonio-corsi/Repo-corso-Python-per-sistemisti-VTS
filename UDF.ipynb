# Funzione vs metodo

Sia le funzioni che i metodi sono **blocchi di codice** <u>richiamabili</u> che accettano una lista di **parametri di input opzionali**, indicati tra <u>parentesi tonde ()</u>.

La differenza sta in come si richiamano.

Una funzione si richiama partendo **dal suo nome**:

```python
lista = [1,2,3,4,5]        # una lista è una collezione di valori (anche di tipo diverso) racchiusi tra parentesi quadre e separati da virgole
print(len(lista))          # la funzione 'len' restituisce il numero di elementi di una lista (o di un'altra collezione)
print("ciao")              # la funzione 'print' visualizza una stringa (o un numero o un'altra cosa) in un formato semplice, senza CSS
```

Un metodo **si applica a un oggetto di una classe**, tramite la [*dot notation*](https://realpython.com/ref/glossary/dot-notation/#:~:text=It%20provides%20a%20way%20to,Python%20code%20intuitive%20and%20readable.):

```python
lista.append(6)
print(lista)
testo = "ciao"
testo.upper()          # restituisce la stringa in maiuscolo
```

Un metodo appartiene a **una classe** (verificabile con la funzione Python nativa `type()`): se un oggetto appartiene a una certa classe, **eredita automaticamente tutti i metodi di quella classe**. Per esempio, ogni oggetto di tipo python `str` può usare i metodi `.upper()`, `.split()`, `.replace()`, perché sono <u>metodi della classe `str`</u>.

Infine, sia le funzioni che i metodi possono essere:
* **nativi** di Python: es. la funzione `len()`, il metodo `.append()`,
* forniti da un **package**: es. `np.array()`, oppure `df.head()`,
* definiti dall'utente (le funzioni con `def`, o i metodi nelle classi che scriviamo noi).

# Le funzioni utente

Nella programmazione tradizionale (detta anche "funzionale") grande importanza è data alle funzioni.

Le funzioni definite dall'utente (*UDF= User Defined Functions*) sono funzioni **create dall'utente ad hoc per qualche scopo**, ad esempio per eseguire operazioni specifiche sui dati. Spesso sono usate per **incapsulare una sequenza di istruzioni e riutilizzarla più volte nel programma**. 

Sono presenti in molti linguaggi e ambienti (Python, SQL, Excel, Spark, ecc.).

In Python, è possibile definire UDF utilizzando **la parola chiave `def`**.

Ecco un esempio di come creare una UDF per calcolare **il volume di una sfera dato il raggio**:

```python
import math

def volume_sfera(raggio):
    """
    Calcola il volume di una sfera dato il raggio.

    Parametri
    ----------
    raggio : float
        Raggio della sfera.

    Restituisce
    -----------
    float
        Volume della sfera.
    """
    return (4 / 3) * math.pi * raggio**3


# Esempio di utilizzo
r = 5
v = volume_sfera(r)

print(f"Volume della sfera di raggio {r}: {v:.2f}")
```

## Caratteristiche principali

**1. Sono create dall'utente**

A differenza delle funzioni predefinite (`len()`, `sum()`, `max()`), vengono scritte dal programmatore.

```python
def volume_sfera(raggio):
    import math
    return (4/3) * math.pi * raggio**3

volume_sfera(5)
```

**2. Possono ricevere parametri in input**

```python
def area_rettangolo(base, altezza):
    return base * altezza
```

**3. Possono restituire un valore**

```python
def quadrato(x):
    return x**2
```

oppure non restituire nulla esplicitamente:

```python
def saluta(nome):
    print(f"Ciao {nome}")
```

**4. Favoriscono la riusabilità del codice**

Invece di scrivere più volte:

```python
volume = (4/3) * math.pi * r**3
```

si può scrivere:

```python
volume = volume_sfera(r)
```

**5. Migliorano leggibilità e manutenzione**

```python
import pandas as pd

df = pd.DataFrame({
    "prezzo": [100, 250, 80, 150],
    "percentuale": [10, 20, 5, 15]
})

def applica_sconto(prezzo, percentuale):
    return prezzo * (1 - percentuale / 100)

# versione vettorizzata
df["prezzo_scontato"] = applica_sconto(df["prezzo"], df["percentuale"])

print(df)
```

**6. Possono essere richiamate più volte**

```python
for r in [1, 2, 3]:
    print(volume_sfera(r))
```

**7. Possono richiamare altre funzioni**

```python
def area_cerchio(r):
    import math
    return math.pi * r**2

def volume_cilindro(r, h):
    return area_cerchio(r) * h
```

**8. Possono avere valori di default**

```python
def potenza(x, esponente=2):
    return x**esponente

print(potenza(5))      # 25
potenza(5, 3)   # 125
```

## Vantaggi

✅ Riutilizzo del codice
✅ Maggiore leggibilità
✅ Riduzione degli errori
✅ Manutenzione più semplice
✅ Modularità del programma

## Esempio in Pandas

Una UDF può essere applicata a una Series:

```python
import pandas as pd

# Creazione del DataFrame
df = pd.DataFrame({
    "età": [12, 25, 17, 42, 8, 19]
})

print(df)

def classifica_eta(x):
    if x < 18:
        return "minore"
    else:
        return "adulto"

df["categoria"] = df["età"].apply(classifica_eta)
```

Tuttavia, in Pandas è spesso preferibile usare operazioni vettorizzate:

```python
import numpy as np

df["categoria"] = np.where(df["età"] < 18,
                           "minore",
                           "adulto")
```

perché risultano molto più veloci delle UDF applicate elemento per elemento.

In sintesi, una **User Defined Function** è una funzione scritta dal programmatore per svolgere uno specifico compito, rendendo il codice più modulare, riutilizzabile e leggibile.

| Aspetto | Funzioni native | Funzioni definite dall'utente (UDF) |
|-----------|----------------|-------------------------------------|
| Chi le crea | Python o una libreria | Il programmatore |
| Disponibilità | Già pronte all'uso | Devono essere definite prima dell'uso |
| Scopo | Operazioni generiche e comuni | Risolvere esigenze specifiche |
| Prestazioni | Generalmente molto ottimizzate | Dipendono dall'implementazione |
| Modificabili | No | Sì |
| Riutilizzabili | Sì | Sì |
| Esempi | `len()`, `sum()`, `max()`, `print()` | `volume_sfera()`, `applica_sconto()` |
| Definizione | Nessuna definizione necessaria | Richiedono l'istruzione `def` |
| Esempio d'uso | `len(lista)` | `applica_sconto(prezzo, percentuale)` |
