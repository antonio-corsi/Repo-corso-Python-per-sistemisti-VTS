# Conflitti tra package e python
Sia i package che python (l'interprete *.exe*) sono **versionati**, e le versioni possono andare in conflitto.

Google Colab ci garantisce **consistenza** tra python e i package installati.

Il problema nasce quando dobbiamo installare un nuovo package che NON è presente nella lista dei package pre-installati.

Facciamo un esempio: il package `dmba`, il package di un libro su python.

```python
!pip show dmba    # non c'è
```

```python
!pip install dmba
```

Ogni package richiede, per funzionare correttamente, che altri package siano installati. Si chiamano **dipendenze**. Se le dipendenze sono **già installate**, *pip install* visualizza questo messaggio:<br>
> *Requirement already satisfied*

Se tutto va bene, l'installazione si conclude con:<br>

> *Installing collected packages: dmba*<br>
> *Successfully installed dmba-0.2.4*

E' possibile, per varie ragioni (legate agli ambienti virtuali), che l'installazione del package dià un warning o addirittura un errore. Non è detto, tuttavia, che l'installazione sia fallita. Il criterio sono le ultime due righe.

```python
!pip install dmba
```

```python
!pip show numpy
```

```python
!pip uninstall numpy
```

```python
!pip install numpy==1.26.0
```

pip segnala che alcuni package NON sono compatibili con la versione 1.2.6.0 e perciò non funzioneranno.

La soluzione sono gli ambienti virtuali (locali al PC) creati con [*conda*](https://anaconda.org/channels/anaconda/packages/conda/overview) oppure [*venv*](https://docs.python.org/3/library/venv.html). conda e venv sono due "gestori di ambienti virtuali python", ma conda è anche un gestore di package (come alternativa a pip). conda (consigliata la versione **miniconda**, molto più comoda della mastodontica versione **anaconda**) è un strumento terzo (non fa parte di Python) e permette di gestire ambienti virtuali anche non in python. venv fa invece parte di Python.
La scelta preferenziale è in genere *conda*.

### Esempi di comandi Magic comuni
Ecco alcuni dei comandi più utili per analizzare le performance e gestire l'ambiente:

`%who` / `%whos`: Elenca le variabili definite nel notebook.

`%timeit`: Misura il tempo di esecuzione di un'istruzione.

`%pwd`: Mostra la cartella di lavoro corrente.

`%ls`: Elenca i file nella cartella corrente.

`%%writefile`: Salva il contenuto della cella in un file.

`%history`: Mostra la cronologia dei comandi eseguiti.


```python
# Misura il tempo di esecuzione di un'operazione
%timeit [x**2 for x in range(1000)]   # for in-line
```
